```python
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period = "max")
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
#webscraping
!pip install requests
!pip install bs4
!pip install html5lib
!pip install lxml
!pip install plotly

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
tesla_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
data = requests.get(tesla_url).text
print(data)
```

```python
#Webscraping for tesla stocks

soup = BeautifulSoup(data, "html.parser")
tesla_revenue = pd.DataFrame(columns=["Year","Revenue"])
for row in soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    year = col[0].text
    revenue = col[1].text[1::].replace(",","") #use slicing to remove dollar sign and replace to remove commas

    tesla_revenue = pd.concat([tesla_revenue,pd.DataFrame({"Year":[year], "Revenue":[revenue]})], ignore_index=True)
tesla_revenue.tail()
```

```python
#gamestop data

gme = yf.Ticker("GME")
gme_data = gme.history(period = "max")
gme_data.reset_index(inplace=True)
gme_data.head()
```

```python
#gamestop revenue with webscraping


gme_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
```

```python
gme_scrap_data = requests.get(gme_url).text
print(gme_scrap_data)
```

```python
gme_soup = BeautifulSoup(gme_scrap_data, "html.parser")
gme_revenue = pd.DataFrame(columns=["Year","Revenue"])
for row in soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    year = col[0].text
    revenue = col[1].text[1::].replace(",","")  #Use slicing to remove dollar sign
    gme_revenue = pd.concat([gme_revenue,pd.DataFrame({"Year":[year], "Revenue":[revenue]})], ignore_index=True)
gme_revenue.tail()
```

```python
#Charts
import plotly.graph_objects as go
from  plotly.subplots import make_subplots
#define make_graph function

def make_graph(stock_data, revenue_data, stock):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Historical Share Price", "Historical Revenue"), vertical_spacing = .3)
    fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data.Date, infer_datetime_format=True), y=stock_data.Close.astype("float"), name="Share Price"), row=1, col=1)
    fig.add_trace(go.Scatter(x=pd.to_datetime(revenue_data.Year, infer_datetime_format=True), y=revenue_data.Revenue.astype("float"), name="Revenue"), row=2, col=1)
    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)
    fig.update_yaxes(title_text="Price ($US)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=1)
    fig.update_layout(showlegend=False,
    height=900,
    title=stock,
    xaxis_rangeslider_visible=True)
    fig.show()


make_graph(tesla_data,tesla_revenue,"Tesla")

```

```python
print(tesla_data)
make_graph(gme_data,gme_revenue,"Gamestop")
```
